```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from glob import glob
```

```python
female_path = glob('./003 data/female/*.jpg')
male_path = glob('./003 data/male/*.jpg')
```

```python
print(f'The number of female images are {len(female_path)}')
print(f'The number of male images are {len(male_path)}')
```

```python
l# Load the model
model = cv2.CascadeClassifier('./models/haarcascade_frontalface_default.xml')
```

```python
for i in range(len(female_path)):
    try:
    # read the image
        img = cv2.imread(female_path[i])

        # Convert the image from BGR to RBG
        img_rbg = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        # Apply the model
        model = cv2.CascadeClassifier('./models/haarcascade_frontalface_default.xml')
        gray_scale_image = cv2.cvtColor(img_rbg, cv2.COLOR_RGB2GRAY)
        face_list = model.detectMultiScale(gray_scale_image, scaleFactor=1.5, minNeighbors=5)
        print(face_list)
        for x,y,w,h in face_list:
            #cv2.rectangle(img_rbg, (x,y), (x+w,y+h), (0,225,0),2)
        #crop the face(Region of intrest)
            roi = img[y:y+h, x:x+w]

            # Save the croped images to the croped_data folder
            cv2.imwrite(f'./croped_data/female/female{i}.jpg',roi)
            print('image sucessfully processed')
    except:
        print('Unable to process the image')
```

```python
for i in range(len(male_path)):
    try:
    # read the image
        img = cv2.imread(male_path[i])

        # Convert the image from BGR to RBG
        img_rbg = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        # Apply the model
       
        gray_scale_image = cv2.cvtColor(img_rbg, cv2.COLOR_RGB2GRAY)
        face_list = model.detectMultiScale(gray_scale_image, scaleFactor=1.5, minNeighbors=5)
        print(face_list)
        for x,y,w,h in face_list:
            
        #crop the face(Region of intrest)
            roi = img[y:y+h, x:x+w]

            # Save the croped images to the croped_data folder
            cv2.imwrite(f'./croped_data/male/male{i}.jpg',roi)
            print('image sucessfully processed')
    except:
        print('Unable to process the image')
```

```python
# Save the croped images to the croped_data folder
plt.imshow(roi)
plt.axis('off')
plt.show()
```
